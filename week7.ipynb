```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns

df=pd.read_csv("C:\\Users\\ABHISHEK KUMAR\\Downloads\\data.csv")
df.head()

#spliting the dataset
x=pd.read_csv("C:\\Users\\ABHISHEK KUMAR\\Downloads\\data.csv")
y=x.pop("diagnosis")
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=42) 

#building the model
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

#accuracy
print("Accuracy score:",accuracy_score(y_test,y_pred))
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,annot_kws={"size":30},fmt="d",cmap="Blues")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.svm import SVR

#loading the dataset
df=pd.read_csv("C:\\Users\\ABHISHEK KUMAR\\Downloads\\data.csv")
df.head()

#spliting dataset
x=df.drop(['diagnosis'],axis='columns').values
y=df["diagnosis"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
y_train=y_train.reshape(-1,1)
y_test=y_test.reshape(-1,1)

#building model
model=SVR(kernel='rbf')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.svm import SVR

data=pd.read_csv("C:\\Users\\ABHISHEK KUMAR\\Downloads\\data.csv")
data.diagnosis=[1 if i=='M' else 0 for i in data.diagnosis]
from sklearn.model_selection import train_test_split

x=data.drop(['diagnosis'],axis=1)
y=data.diagnosis.values

x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.2,random_state=42)

mo=RandomForestClassifier()
mo.fit(x_tr,y_tr)
y_pr=mo.predict(x_ts)

m=SVR(kernel='rbf')
m.fit(x_tr,y_tr)
print(m.score(x_ts,y_ts))


print("Accuracy Score:",accuracy_score(y_ts,y_pr))
cm=confusion_matrix(y_ts,y_pr)
sns.heatmap(cm,annot=True,fmt="d",cmap="Greens")
```
