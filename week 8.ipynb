```python
import numpy as np
import pandas as pd
import seaborn as s
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC

data=pd.read_csv("C:\\Users\\ABHISHEK KUMAR\\Downloads\\apples_and_oranges (1).csv")
data

import seaborn as s
s.scatterplot(x='Weight',y='Size',hue='Class',data=data)

tr_set,tst_set=train_test_split(data,test_size=0.2,random_state=1)
print("Train Set:",tr_set)
print("\nTest Set:",tst_set)

x_tr=tr_set.iloc[:,0:2].values
y_tr=tr_set.iloc[:,2].values
x_ts=tst_set.iloc[:,0:2].values
y_ts=tst_set.iloc[:,2].values
print(x_tr,y_tr)
print(x_ts,y_ts)

cl=SVC(kernel='rbf',random_state=1,C=1,gamma='auto')
cl.fit(x_tr,y_tr)

y_pr=cl.predict(x_ts)
print(y_pr)

cm=confusion_matrix(y_ts,y_pr)
print(cm)
acc=float(cm.diagonal().sum())/len(y_ts)
print("Model accuracy is:",acc*100,'%')

```

```python
import pandas as p
import seaborn as s
import matplotlib.pyplot as m
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import r2_score,confusion_matrix
from sklearn.preprocessing import LabelEncoder

da=p.read_csv("C:\\Users\\ABHISHEK KUMAR\\Downloads\\fish_prediction.csv")
print(da['Species'].value_counts())
s.pairplot(data=da,hue='Species')
m.show()

label_encoder = LabelEncoder()
da['Species'] = label_encoder.fit_transform(da['Species'])

x=da.drop(['Species'],axis=1)
y=da['Species']
xtra,xtes,ytra,ytes=train_test_split(x,y,test_size=0.2,random_state=2)

mod=SVC(kernel='rbf',C=1,gamma='auto')
mod.fit(xtra,ytra)
ypred=mod.predict(xtes)
cm=confusion_matrix(ytes,ypred)
s.heatmap(cm,annot=True,fmt='d',cmap="viridis")
m.title("Confusion matrics ")
m.show()
print('Accuracy:',r2_score(ytes,ypred))
```
